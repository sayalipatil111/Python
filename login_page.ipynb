```python
uname=["Admin","Admin1","Admin2","Admin3","Admin4"]
passwd=["12345","5678","111","222","333","444"]


c=1

while c<=3:
    
    user_name=input("Enter user name or mobile no or email : ")
    password=input("Enter password  : ")

    x=0
    while x<=4:
        
        if(uname[x]==user_name and passwd[x]==password):
            
            print("Login Sucessfull...!!")
            break
    
        else:
            print("Invalid userid or password")
            break
        x=x+1
    c=c+1
```
